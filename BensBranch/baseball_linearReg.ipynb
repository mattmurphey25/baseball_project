```python
import pandas as pd  # to store tabular data

baseball = pd.read_csv('data_union.csv')
print(baseball.columns)
```

```python
baseball.head(30)
```

```python
selected_years = baseball[ baseball['Year'] != 2018 ]
print(selected_years.head(10))
print(selected_years.tail(10))
```

```python
# Ignore year and team name
# Ignore losses since that has direct correlation to wins
# Ignore Runs/Game since that has direct correlation to runs
# Ignore Earned Runs since that direct correlation to ERA
selected_columns = ['Runs', 'Hits', 'Stolen_Bases', 'BA',
      'Struck_Out', 'TB', 'SV', 'Hits_Allowed', 'Runs_Allowed', 'Strike_Outs',
      'ERA', 'Rtot', 'DefEff']
X = selected_years[selected_columns]
X.head(30)
```

```python
y = selected_years['W']
y.head(30) #2008
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2018)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

# Show single column histogram
#plt.hist(baseball['ERA'])
#plt.show()

# Show all columns' histograms
X.hist(figsize=(4, 4), sharex=False)
plt.show()
```

```python
# Above plots look mostly normal so try StandardScaler as first preprocessing step

from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler, MinMaxScaler
#from sklearn.linear_model import LinearRegression
#regressor = LinearRegression()
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
pipeline = make_pipeline(MinMaxScaler(), regressor)
pipeline.fit(X_train, y_train)
```

```python
# Get R^2 for this one train/test split in one step
r_squared = pipeline.score(X_test, y_test)
print("r_squared = {}".format(r_squared))
```

```python
print(regressor.coef_)
```

```python
print(regressor.intercept_)
```

```python
# Or alternatively from using prediction and score separatly
from sklearn.metrics import mean_squared_error, r2_score
y_pred = pipeline.predict(X_test)
print(mean_squared_error(y_test, y_pred))
print(r2_score(y_test, y_pred))
```

```python
for year in range(2008, 2018):
    winner_stats = baseball[ (baseball['Year'] == year) & (baseball['Winner'] == 1)] [selected_columns]
    print(year, '\n', winner_stats, '\n\n')
```

```python
for year in range(2008, 2018):
    winner_stats = baseball[ (baseball['Year'] == year) & (baseball['Winner'] == 1)] [selected_columns]
    winner_names = baseball[ (baseball['Year'] == year) & (baseball['Winner'] == 1)] ["Tm"]
    result = pipeline.predict(winner_stats.values)
    print(year, result, winner_names)
```

```python
# Try cross-validation to select different folds of training and testing sets
from sklearn.model_selection import cross_val_score
scores = cross_val_score(pipeline, X, y, cv=10, scoring='accuracy')
print(scores)
print(scores.mean())
```
