```python
def myColor(winner):
   if winner == 0: return "red"
   return "green"

data_union["color"] = data_union["Winner"].apply(myColor)
```

```python
data_union.plot(kind="scatter", x= "Runs", y="Hits", grid=True, figsize=(20,10), title="Hits vs. Runs", marker="o", c=data_union["color"])
plt.show()
```

```python
data_union.plot(kind="scatter", x= "W", y="BA", grid=True, figsize=(20,10), title="Wins vs. Batting Average", marker="o", c=data_union["color"])
plt.show()
```

```python
data_union.plot(kind="scatter", x= "W", y="R/G", grid=True, figsize=(20,10), title="W vs. R/G", marker="o", c=data_union["color"])
plt.show()
```

```python
data_union.plot(kind="scatter", x= "Strike Outs", y="DefEff", grid=True, figsize=(20,10), title="Strike Outs vs. Defensive Efficiency", marker="o", c=data_union["color"])
plt.show()
```
