```python
import pandas as pd  # to store tabular data

baseball = pd.read_csv('data_union.csv')
print(baseball.columns)
```

```python
baseball.head(30)
```

```python
selected_years = baseball[ baseball['Year'] != 2018 ]
print(selected_years.head(10))
print(selected_years.tail(10))
```

```python
# Ignore year and team name
# Ignore losses since that has direct correlation to wins
# Ignore Runs/Game since that has direct correlation to runs
# Ignore Earned Runs since that direct correlation to ERA
selected_columns = ['W', 'Runs', 'Hits', 'Stolen_Bases', 'BA',
      'Struck_Out', 'TB', 'SV', 'Hits_Allowed', 'Runs_Allowed', 'Strike_Outs',
      'ERA', 'Rtot', 'DefEff']
X = selected_years[selected_columns]
X.head(30)
```

```python
y = selected_years['Winner']
y.head(30) #2008
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2018)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

# Show single column histogram
#plt.hist(baseball['ERA'])
#plt.show()

# Show all columns' histograms
X.hist(figsize=(4, 4), sharex=False)
plt.show()
```

```python
# Above plots look mostly normal so try StandardScaler as first preprocessing step

from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
tree_clf = DecisionTreeClassifier(random_state=1999)
pipeline = make_pipeline(MinMaxScaler(), tree_clf)
pipeline.fit(X_train, y_train)
```

```python
# Get accuracy for this one train/test split in one step
accuracy = pipeline.score(X_test, y_test)
print("accuracy = {}".format(accuracy))
```

```python
# Or alternatively from using prediction and score separatly
from sklearn import metrics
y_pred = pipeline.predict(X_test)
print(metrics.accuracy_score(y_test, y_pred))
```

```python
# What are most important features?
sorted(zip(tree_clf.feature_importances_, selected_columns), reverse=True)
```
