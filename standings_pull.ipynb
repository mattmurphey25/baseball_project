```python
import pandas as pd
```

```python
year = 2017
url = f"https://www.baseball-reference.com/leagues/MLB/{year}-standings.shtml"
tables = pd.read_html(url)
```

```python
frames = [tables[0],tables[1],tables[2],tables[3],tables[4],tables[5]]
full_df = pd.concat(frames)
full_df.sort_values(by='W', ascending=False)
full_df = full_df.reset_index(drop=True)
full_df
```

```python
full_df.to_csv(f"Resources/{year}_standings.csv")
```
