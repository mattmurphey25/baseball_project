```python
full_test_df = full_test_df.reset_index(drop=True)
full_test_df.head()
```

```python
train_value = []
for row in full_train_df["WAR"]:
    if row > 70:
        train_value.append("70+ WAR")
    elif row > 40:
        train_value.append("40<WAR<70")
    elif row > 20:
        train_value.append("20<AR<40")
    elif row > 10:
        train_value.append("10<WAR<20")
    elif row > 5:
        train_value.append("5<WAR<10")
    elif row > 0:
        train_value.append("0<WAR<5")
    elif row == 0:
        train_value.append("NO WAR")
    elif row < 0:
        train_value.append("<0 WAR")
```

```python
data = {"WAR": train_value}
train_WAR_df = pd.DataFrame(data)
train_WAR_df.head()
```

```python
del full_train_df["WAR"]
del full_test_df["WAR"]
train_data_binary_encoded = pd.get_dummies(full_train_df)
test_data_binary_encoded = pd.get_dummies(full_test_df)
```

```python
output_value = len(train_WAR_df.WAR.unique())
```

```python
y_train = train_WAR_df
```

```python
X_train = train_data_binary_encoded

X_test = test_data_binary_encoded
print("Train shape: ", X_train.shape, y_train.shape)
print("Test shape: ", X_test.shape)
```

```python
X_train = pd.read_csv("Resources/X_train_Rnd_OvPck.csv")
X_test = pd.read_csv("Resources/X_test_Rnd_OvPck.csv")
```

```python
print("Train shape: ", X_train.shape, y_train.shape)
print("Test shape: ", X_test.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense

# Create model and add layers
def build_classifier():
    classifier = Sequential()
    classifier.add(Dense(units=200, activation='relu', input_dim=190))
    classifier.add(Dense(units=100, activation='relu'))
    classifier.add(Dense(units=100, activation='relu'))
    classifier.add(Dense(units=output_value, activation='softmax'))
    classifier.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return classifier

# Create a Keras model that's compatible with scikit-learn
from keras.callbacks import EarlyStopping
from keras.wrappers.scikit_learn import KerasClassifier
keras_classifier = KerasClassifier(build_classifier, epochs=300, shuffle=True, verbose=2, callbacks=[EarlyStopping(monitor='acc', patience=20, verbose=2)])
```

```python
from sklearn.preprocessing import LabelEncoder
from keras.utils import to_categorical

# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)

# Step 2: Convert encoded labels to one-hot-encoding
y_train_categorical = to_categorical(encoded_y_train)
```

```python
# Ensure we are scaling features
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
model = make_pipeline(StandardScaler(), keras_classifier)
```

```python
model.fit(X_train, y_train_categorical)
```

```python
# Save the model
keras_classifier.model.save("draft_model_trained_rnd_ovpck.h5")
```

```python
encoded_predictions = model.predict(X_test)
encoded_predictions
```

```python
prediction_labels = label_encoder.inverse_transform(encoded_predictions)
```

```python
output_df["WAR_prediction"] = prediction_labels
```
